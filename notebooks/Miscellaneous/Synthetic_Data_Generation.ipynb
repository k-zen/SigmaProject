```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
DATA_POINTS = 100 # Number of points in the X axis to generate.
RANGE = (-5, 5) # Low & High

# x = np.linspace(RANGE[0], RANGE[1], DATA_POINTS) # Linearly spaced points.
x = np.random.uniform(low=RANGE[0], high=RANGE[1], size=DATA_POINTS)

y = 3 * (x ** 3)
```

```python
plt.plot(x, y, 'r')
plt.show()
```

```python
tmp_df = pd.DataFrame(
    data={
        'X': list(x),
        'Y': y
    }
)
tmp_df.to_csv(path_or_buf='/Users/akc/Downloads/data.csv')
```
